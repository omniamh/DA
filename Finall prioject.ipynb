<h1 style="color:#A0A0A0; font-size:18px; text-align:center; font-weight:bold; text-align:left;">

 5-Checking & Understanding Our Data

</h1>

```python
data = data.dropna(subset=['country'])
```

```python
data = data.dropna(subset=['company'])
```

```python
data = data.dropna(subset=['children'])
```

```python
data['agent'] = data['agent'].fillna(0)
```

```python
data.duplicated().sum()
```

```python
data = data.drop_duplicates()
```

```python
data.dtypes
```

<hr style="border: 1px solid #62B5E5;">
<h1 style="color:#A0A0A0; font-size:18px; text-align:center; font-weight:bold; text-align:left;">

 6-Creating the Date column

</h1>

```python
data['Date'] = pd.to_datetime(
                       data['arrival_date_year'].astype(str) + '-' +
                       data['arrival_date_month'].astype(str) + '-' +
                       data['arrival_date_day_of_month'].astype(str),
                       format='%Y-%B-%d'
)
data = data.drop(columns=['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month'])
```

<h1 style="color:#A0A0A0; font-size:18px; text-align:center; font-weight:bold; text-align:left;">

 7-Converting Object columns to String

</h1>

```python
data['hotel'] = data['hotel'].astype(str)
data['meal'] = data['meal'].astype(str)
data['country'] = data['country'].astype(str)
data['market_segment'] = data['market_segment'].astype(str)
data['distribution_channel'] = data['distribution_channel'].astype(str)
data['reserved_room_type'] = data['reserved_room_type'].astype(str)
data['reservation_status'] = data['reservation_status'].astype(str)
data['deposit_type'] = data['deposit_type'].astype(str)
data['assigned_room_type'] = data['assigned_room_type'].astype(str)
data['reservation_status_date'] = data['reservation_status_date'].astype(str)
```
